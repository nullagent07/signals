### Загрузка исторических данных по каждому символу

```python
limit = 2280
candles = '15m'

for symbol in usdt_swap_symbols:
    # Формирование пути для сохранения файлов
    directory_path = f'./data/{symbol.replace("/", "_")}/Candles_{candles}'
    file_path = os.path.join(directory_path, f'{symbol.replace("/", "_")}_{limit}_{candles}.csv')

    # Создание директорий, если они не существуют
    os.makedirs(directory_path, exist_ok=True)
    
    # Загрузка последних свечей
    ohlcv = exchange.fetch_ohlcv(symbol, candles, limit=limit)
    
    # Преобразование данных в DataFrame
    df = pd.DataFrame(ohlcv, columns=['timestamp', 'open', 'high', 'low', 'close', 'volume'])

    # Преобразование timestamp в читаемый формат
    df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms')

    # Сохранение данных в CSV файл
    # df.to_csv(f'./historical_data_{symbol}.csv', index=False)
    
    df.to_csv(file_path, index=False)
```
